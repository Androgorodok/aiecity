Импорт библиотеки numpy

Создание единичной матрицы 4 на 4 с элементами типа int64, которые умножилимь на 6

```python
import numpy as np
a=np.eye(4, dtype='int64')*6
print(a)
```

Размерность матрицы

```python
a.shape
```

Создание нулевой матрицы размером 4 на 4 с элементами типа int64

```python
b=np.zeros((4, 4), dtype='int64')
b
```

Создание матрицы с элементами от 0 до 10000 с шагом 1000

```python
c=np.arange(0, 10000+1, 1000)
c
```

Элементы матрицы от 2 до 12 с шагом 3 равны 6

```python
c[1:12:3]=6
c
```

Умножение матриц

Операция выполнилась успешно, так как размеры матриц позволяют их умножить

```python
a*b
```

Создание единичной матрицы 4 на 4 типа float64

Все элементы матрицы стали равны косинусу своего значения

```python
d=np.eye(4, dtype="float64")
d[::]=np.cos(a)
d
```

Округлить значения матрицы до 2 знаков после запятой

```python
np.round(d, 2)
```

Округлить элементы матрицы в большую сторону

```python
np.ceil(d)
```

Округлить элементы матрицы в меньшую сторону

```python
np.fix(d)
```

Элементы каждой из 4 матриц были просуммированы между собой, а затем суммы элементов этих матриц были сложены между собой

```python
k=np.sum(a)+np.sum(b)+np.sum(c)+np.sum(d)
k
```

Умножение и сложение матриц

```python
a*d+d
```

Вектор-столбец стал вектор-строкой через операцию транспонирования
